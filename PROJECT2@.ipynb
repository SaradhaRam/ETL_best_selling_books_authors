```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "../Project_2/NYT_bestsellers_2011_2018.csv"
new_york_times_data = pd.read_csv(csv_file,encoding='ISO-8859-1')
new_york_times_data.head()
```

```python
new_york_times_dt = new_york_times_data.drop(['dagger', 'book_review_link', "primary_isbn13","sunday_review_link","date","first_chapter_link","contributor","weeks_on_list","price"], axis=1)
```

```python
final_new_york_df=new_york_times_dt.dropna(subset=["primary_isbn10"], how ="any")
final_new_york_df
```

```python
rr = final_new_york_df.rename(columns={"primary_isbn10":"ISBN-10"})
final_newyork_data = rr.loc[rr["ISBN-10"]!="None"]
final_newyork_data = final_newyork_data[["ISBN-10","title","author","publisher","age_group"]]
final_newyork_data
```
